# Datos meteorológicos | IVIA

```python
import requests
import pandas as pd
from time import sleep

def get_meteo_data_por_año(año_inicio, año_fin):
    url = "http://riegos.ivia.es/ajax/calculos.php"
    headers = {
        'Content-Type': 'application/x-www-form-urlencoded',
        'Referer': 'http://riegos.ivia.es/datos-meteorologicos'
    }
    
    columnas_utiles = ['fecha', 'hora', 'tmedia', 'hmedia', 'radiacion', 'viento',
                       'direccion', 'direccion_num', 'precipitacion', 'sestado']

    todos_los_datos = []

    for año in range(año_inicio, año_fin + 1):
        print(f"Descargando datos de {año}...")

        data = {
            't': 'meteorologia',
            'r': 'semihorarios',
            'e[0][p]': '46',
            'e[0][e]': '30',
            'i[y]': str(año),
            'i[m]': '3',
            'i[d]': '1',
            'f[y]': str(año),
            'f[m]': '11',
            'f[d]': '31',
        }

        response = requests.post(url, data=data, headers=headers)
        response.raise_for_status()
        json_data = response.json()
        df = pd.DataFrame(json_data['data'])
        df['fecha'] = pd.to_datetime(df['fecha'], dayfirst=True)
        df = df[columnas_utiles]
        todos_los_datos.append(df)
        
        sleep(1)  

    df_final = pd.concat(todos_los_datos, ignore_index=True)
    df_final.set_index('fecha', inplace=True)
    print('Descarga completada')
    return df_final

# Ejecutar
df_meteo_completo = get_meteo_data_por_año(2020,2023)
```
